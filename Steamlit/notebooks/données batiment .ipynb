```python

# """ Suppression colonnes jugées peu pertinentes pour la suite des analyses   """


def drop_columns(df, columns_to_drop):
    """
    Supprime les colonnes spécifiées d'un DataFrame.
    
    Args:
    - df : DataFrame : Le DataFrame à modifier.
    - columns_to_drop : list : Liste des noms des colonnes à supprimer.
    
    Returns:
    - df : DataFrame : Le DataFrame avec les colonnes supprimées.
    """
    return df.drop(columns_to_drop, axis=1)

# """ Liste des colonnes à supprimer """
columns_to_drop = ['Invariant_fiscal_logement','N°_immatriculation_copropriété','Nom__commune_(Brut)',"Complément_d'adresse_bâtiment",'Code_postal_(BAN)',"Complément_d'adresse_logement","Complément_d'adresse_logement",'Nom_résidence','Nom__rue_(BAN)','Adresse_(BAN)','N°_voie_(BAN)','Adresse_brute','Adresse_brute','N°DPE', 'Date_réception_DPE', 'Date_établissement_DPE', 'Date_visite_diagnostiqueur', 'Modèle_DPE', 'N°_DPE_remplacé', 'Date_fin_validité_DPE', 'Version_DPE', 'N°_DPE_immeuble_associé', 'Méthode_du_DPE']

# """ Appel de la fonction pour supprimer les colonnes """
df = drop_columns(df, columns_to_drop)

# """ Affichage du DataFrame résultant """
pd.set_option('display.max_columns', None)
# Remplacer les NaN par -1 par exemple
df['Année_construction'].fillna(0, inplace=True)

df

```

## Imputation par la moyenne 

```python
def fillna_with_mean(df, columns):
    """
    Remplace les valeurs manquantes dans les colonnes spécifiées par la moyenne de chaque colonne.

    Args:
    - df : DataFrame : Le DataFrame à modifier.
    - columns : list : Liste des noms des colonnes à traiter.

    Returns:
    - df : DataFrame : Le DataFrame avec les valeurs manquantes remplacées par la moyenne.
    """
    for column in columns:
        df[column] = df[column].fillna(df[column].mean())
    return df

# Liste des colonnes à traiter
columns_to_fill = ['Conso_kWhep/m²/an','Emission_GES_kgCO2/m²/an','Conso_é_finale_énergie_n°1','Conso_é_primaire_énergie_n°1','Frais_annuel_énergie_n°1','Année_relève_conso_énergie_n°1','Conso_é_finale_énergie_n°2','Conso_é_primaire_énergie_n°2','Frais_annuel_énergie_n°2','Année_relève_conso_énergie_n°2','Conso_é_finale_énergie_n°3','Conso_é_primaire_énergie_n°3','Frais_annuel_énergie_n°3','Année_relève_conso_énergie_n°3']

# """Appel de la fonction pour remplacer les valeurs manquantes par la moyenne"""
df = fillna_with_mean(df, columns_to_fill)

# Remplacer les valeurs NaN par la valeur la plus fréquente
df['Cage_d\'escalier'].fillna(df['Cage_d\'escalier'].mode()[0], inplace=True)
# """Affichage du DataFrame résultant"""
df
```

```python

# Filtrer les lignes où le type d'énergie est soit 'électricité' soit 'gaz naturel'
df = df[df['Type_énergie_n°1'].isin(['Électricité', 'Gaz naturel'])]
# Maintenant df contient uniquement les lignes avec les types d'énergie spécifiés
df['Catégorie_ERP'].fillna(df['Catégorie_ERP'].mode()[0], inplace=True)
# 'imputation des valeurs manquantes avec la moyenne
df['Nombre_occupant'].fillna(df['Nombre_occupant'].mean(), inplace=True)
df['Surface_(SHON)'].fillna(df['Surface_(SHON)'].mean(), inplace=True)

df
```

# Interpretations


La forte corrélation entre la consommation d'énergie et l'émission de gaz à effet de serre (GES) est assez intuitive, car une grande partie de la consommation d'énergie dans les bâtiments est liée à l'utilisation de sources d'énergie qui émettent des gaz à effet de serre lors de leur combustion. Par exemple, l'utilisation de combustibles fossiles tels que le gaz naturel ou le fioul pour le chauffage ou la production d'électricité peut contribuer à la fois à une consommation élevée d'énergie et à des émissions de GES élevées.

En France, le chauffage des bâtiments est souvent alimenté par le gaz naturel, le fioul, l'électricité et dans une moindre mesure, le bois. Le gaz naturel et le fioul sont des sources d'énergie fossiles qui émettent des GES lors de leur combustion, et ils peuvent donc contribuer à la fois à une consommation élevée d'énergie et à des émissions de GES élevées. Cependant, la France a également investi dans des sources d'énergie à faible teneur en carbone, telles que l'énergie nucléaire et les énergies renouvelables, comme l'énergie éolienne et solaire, ce qui contribue à réduire les émissions de GES dans certains secteurs.

Quant à la corrélation entre la consommation d'énergie et les frais annuels d'énergie pour les niveaux 2 et 3, cela pourrait être dû au fait que les bâtiments qui consomment plus d'énergie ont tendance à avoir des coûts énergétiques plus élevés. Ces coûts incluent souvent non seulement la consommation d'énergie elle-même, mais aussi d'autres frais associés à l'entretien, au chauffage, à la climatisation, etc. Ainsi, les bâtiments avec une consommation d'énergie élevée auront probablement des frais annuels d'énergie plus élevés.

Bien que ces variables soient fortement corrélées, il pourrait y avoir d'autres facteurs sous-jacents qui influencent à la fois la consommation d'énergie et les émissions de GES ou les frais annuels d'énergie

Type de chauffage : Les systèmes de chauffage utilisant des combustibles fossiles comme le gaz naturel ou le fioul ont tendance à entraîner à la fois une consommation d'énergie plus élevée et des émissions de GES plus élevées par rapport aux systèmes de chauffage électriques ou aux énergies renouvelables.

Isolation et étanchéité du bâtiment : Les bâtiments mal isolés ou présentant des fuites d'air auront tendance à nécessiter plus d'énergie pour le chauffage et la climatisation, ce qui pourrait également entraîner des émissions de GES plus élevées et des frais énergétiques plus importants.

Taille et utilisation du bâtiment : Les bâtiments plus grands ou ceux utilisés pour des activités qui nécessitent une utilisation intensive de l'énergie, comme les bâtiments commerciaux ou industriels, auront probablement une consommation d'énergie plus élevée et des frais annuels d'énergie plus importants.

Âge et état du bâtiment : Les bâtiments plus anciens, qui peuvent ne pas être équipés des technologies les plus récentes en matière d'efficacité énergétique, pourraient avoir une consommation d'énergie plus élevée et des frais énergétiques plus importants en raison de la nécessité de rénovations ou de mises à niveau.

Comportement des occupants : Les habitudes des occupants, telles que la température de consigne pour le chauffage et la climatisation, l'utilisation d'appareils électriques, et la gestion de l'éclairage, peuvent influencer de manière significative la consommation d'énergie et les coûts associés.

Tarification de l'énergie : Les différences de coûts de l'énergie entre les régions ou les pays peuvent également influencer les choix énergétiques et les comportements des occupants.

Taille et conception du logement : La taille, la conception et la disposition des logements peuvent avoir un impact significatif sur la consommation d'énergie. Par exemple, les logements mal isolés ou disposant de fenêtres inefficaces peuvent entraîner des pertes de chaleur importantes en hiver et des gains de chaleur excessifs en été, ce qui nécessite une surconsommation d'énergie pour le chauffage et la climatisation.

Comportement des occupants : Les habitudes et le comportement des occupants ont un impact majeur sur la consommation d'énergie des logements. Par exemple, la température de consigne choisie pour le chauffage et la climatisation, la fréquence et la durée de fonctionnement des appareils électriques, l'utilisation de l'eau chaude, etc., peuvent tous influencer la quantité d'énergie consommée.

Équipements et appareils électriques : Les logements sont équipés de divers appareils électriques tels que les réfrigérateurs, les lave-linge, les chauffe-eau, les systèmes de chauffage et de refroidissement, etc. L'efficacité énergétique de ces équipements ainsi que leur utilisation par les occupants peuvent avoir un impact significatif sur la consommation d'énergie globale du logement.

Localisation géographique : La localisation géographique du logement, notamment le climat local, peut influencer la consommation d'énergie. Par exemple, les logements situés dans des climats froids nécessiteront plus d'énergie pour le chauffage, tandis que ceux situés dans des climats chauds nécessiteront plus d'énergie pour la climatisation.

Normes et réglementations : Les normes de construction et les réglementations en matière d'efficacité énergétique peuvent également influencer la consommation d'énergie des logements. Les logements conformes à des normes élevées d'efficacité énergétique peuvent avoir une consommation d'énergie plus faible que ceux qui ne le sont pas.
